```python
import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statistics
```

```python
import warnings


warnings.simplefilter("ignore")

plt.style.use("seaborn-whitegrid")
plt.rc(
    "figure",
    autolayout=True,
    figsize=(11, 4),
    titlesize=18,
    titleweight='bold',
)
plt.rc(
    "axes",
    labelweight="bold",
    labelsize="large",
    titleweight="bold",
    titlesize=16,
    titlepad=10,
)
plot_params = dict(
    color="0.75",
    style=".-",
    markeredgecolor="0.25",
    markerfacecolor="0.25",
    legend=False,
)

%config InlineBackend.figure_format = 'retina'
```

```python
# 2023-03-24
df_20230324 = pd.read_excel('./data/收益率类型_py.xlsx', sheet_name='2023-03-24', index_col=0)
df_20230324
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(df_20230324.index, df_20230324['Spot Rate'], label="Spot Rate")
ax.plot(df_20230324.index, df_20230324['Yield to Maturity'], label="Yield to Maturity")
ax.plot(df_20230324.index, df_20230324['Forward Rate'], label="Forward Rate")


ax.set_title("Yield Curve 2023-03-24")
ax.set_xlabel("Tenor")
ax.set_ylabel("Yield(%)")
fig.autofmt_xdate()
plt.tight_layout()
plt.legend(loc=2)
```
