```python
import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statistics

%matplotlib inline

plt.style.use('fivethirtyeight')

import warnings
warnings.filterwarnings('ignore')
```

```python
# https://www.federalreserve.gov/default.htm
# 美国国债

FRB_H15_df = pd.read_csv('./data/1962-01-02_2023-03-09_FRB_H15.csv', header=0)
FRB_H15_df
```

```python
FRB_H15_df.dtypes
```

```python
FRB_H15_df = FRB_H15_df.replace({"ND": np.nan})
FRB_H15_df = FRB_H15_df.rename(columns={"Time Period": "Time_Period"})
```

```python
FRB_H15_df['Time_Period'] = pd.to_datetime(FRB_H15_df['Time_Period'])
FRB_H15_df['RIFLGFCM01_N.B'] = FRB_H15_df['RIFLGFCM01_N.B'].astype(float)
FRB_H15_df['RIFLGFCM03_N.B'] = FRB_H15_df['RIFLGFCM03_N.B'].astype(float)
FRB_H15_df['RIFLGFCM06_N.B'] = FRB_H15_df['RIFLGFCM06_N.B'].astype(float)
FRB_H15_df['RIFLGFCY01_N.B'] = FRB_H15_df['RIFLGFCY01_N.B'].astype(float)
FRB_H15_df['RIFLGFCY02_N.B'] = FRB_H15_df['RIFLGFCY02_N.B'].astype(float)
FRB_H15_df['RIFLGFCY03_N.B'] = FRB_H15_df['RIFLGFCY03_N.B'].astype(float)
FRB_H15_df['RIFLGFCY05_N.B'] = FRB_H15_df['RIFLGFCY05_N.B'].astype(float)
FRB_H15_df['RIFLGFCY07_N.B'] = FRB_H15_df['RIFLGFCY07_N.B'].astype(float)
FRB_H15_df['RIFLGFCY10_N.B'] = FRB_H15_df['RIFLGFCY10_N.B'].astype(float)
FRB_H15_df['RIFLGFCY20_N.B'] = FRB_H15_df['RIFLGFCY20_N.B'].astype(float)
FRB_H15_df['RIFLGFCY30_N.B'] = FRB_H15_df['RIFLGFCY30_N.B'].astype(float)
FRB_H15_df
```

```python
FRB_H15_df['Time_Period_YM'] = FRB_H15_df['Time_Period'].apply(lambda x: x.replace(day=1))
FRB_H15_df['Time_Period_YM']
```

```python
# 选择每个月月末数据
end_month_sample_df=FRB_H15_df.groupby('Time_Period_YM')['Time_Period', 'RIFLGFCM01_N.B', 'RIFLGFCM03_N.B', 'RIFLGFCM06_N.B', 'RIFLGFCY01_N.B',\
        'RIFLGFCY02_N.B', 'RIFLGFCY03_N.B', 'RIFLGFCY05_N.B', 'RIFLGFCY07_N.B', 'RIFLGFCY10_N.B',\
        'RIFLGFCY20_N.B', 'RIFLGFCY30_N.B'].last()
end_month_sample_df['Time_Period_YM'] = end_month_sample_df.index
end_month_sample_df = end_month_sample_df.reset_index(drop=True)
end_month_sample_df
```

```python
# 计算利差: margin
end_month_sample_df['MARGIN_RIFLGFCM01_N.B'] =end_month_sample_df['RIFLGFCM01_N.B'] - end_month_sample_df['RIFLGFCM01_N.B'].shift(1)
end_month_sample_df['MARGIN_RIFLGFCM03_N.B'] =end_month_sample_df['RIFLGFCM03_N.B'] - end_month_sample_df['RIFLGFCM03_N.B'].shift(1)
end_month_sample_df['MARGIN_RIFLGFCM06_N.B'] =end_month_sample_df['RIFLGFCM06_N.B'] - end_month_sample_df['RIFLGFCM06_N.B'].shift(1)
end_month_sample_df['MARGIN_RIFLGFCY01_N.B'] =end_month_sample_df['RIFLGFCY01_N.B'] - end_month_sample_df['RIFLGFCY01_N.B'].shift(1)
end_month_sample_df['MARGIN_RIFLGFCY02_N.B'] =end_month_sample_df['RIFLGFCY02_N.B'] - end_month_sample_df['RIFLGFCY02_N.B'].shift(1)
end_month_sample_df['MARGIN_RIFLGFCY03_N.B'] =end_month_sample_df['RIFLGFCY03_N.B'] - end_month_sample_df['RIFLGFCY03_N.B'].shift(1)
end_month_sample_df['MARGIN_RIFLGFCY05_N.B'] =end_month_sample_df['RIFLGFCY05_N.B'] - end_month_sample_df['RIFLGFCY05_N.B'].shift(1)
end_month_sample_df['MARGIN_RIFLGFCY07_N.B'] =end_month_sample_df['RIFLGFCY07_N.B'] - end_month_sample_df['RIFLGFCY07_N.B'].shift(1)
end_month_sample_df['MARGIN_RIFLGFCY10_N.B'] =end_month_sample_df['RIFLGFCY10_N.B'] - end_month_sample_df['RIFLGFCY10_N.B'].shift(1)
end_month_sample_df['MARGIN_RIFLGFCY20_N.B'] =end_month_sample_df['RIFLGFCY20_N.B'] - end_month_sample_df['RIFLGFCY20_N.B'].shift(1)
end_month_sample_df['MARGIN_RIFLGFCY30_N.B'] =end_month_sample_df['RIFLGFCY30_N.B'] - end_month_sample_df['RIFLGFCY30_N.B'].shift(1)
```

```python
start_date_str = "1985-01-1"
end_date_str = "2008-12-31";

start_date = datetime.datetime.strptime(start_date_str, "%Y-%m-%d")
end_date = datetime.datetime.strptime(end_date_str, "%Y-%m-%d")
print(start_date, end_date)
```

```python
sample_df = end_month_sample_df[(end_month_sample_df['Time_Period'] >=start_date) 
                                & (end_month_sample_df['Time_Period'] <= end_date)].reset_index(drop=True)
sample_df
```

```python
sample_df
```

```python
# http://nanshu.wang/post/2015-03-15/
# 计算lags阶以内的自相关系数，返回lags个值，将序列均值、标准差视为不变
def autocorrelation(x,lags):#计算lags阶以内的自相关系数，返回lags个值，将序列均值、标准差视为不变
    n = len(x)
    x = np.array(x)
    variance = x.var()
    x = x-x.mean()
    result = np.correlate(x, x, mode = 'full')[-n+1:-n+lags+1]/(variance*(np.arange(n-1,n-1-lags,-1)))
    return result[lags-1]

test = sample_df['RIFLGFCM06_N.B']
#N = len(test)
#np.correlate(test[0:N-1], test[1:N])
autocorrelation(test, 12)
```

```python
# 选择期限6月(RIFLGFCM06_N.B), 12月(RIFLGFCY01_N.B), 24月(RIFLGFCY02_N.B), 36月(RIFLGFCY03_N.B)
# 60月(RIFLGFCY05_N.B), 120月(RIFLGFCY10_N.B)
# 计算： 样本均值， 样本标准差，第一阶和第十二阶自相关系数

import statistics 
calculate_name_list = ['RIFLGFCM06_N.B', 'RIFLGFCY01_N.B', 'RIFLGFCY02_N.B', 'RIFLGFCY03_N.B', 'RIFLGFCY05_N.B', 'RIFLGFCY10_N.B']
N = sample_df.shape[0]
for calculate_name in calculate_name_list:
    sample_mean = sample_df[calculate_name].mean()
    sample_std = sample_df[calculate_name].std() * np.sqrt(N)/np.sqrt(N-1)
    sample_1_lags = autocorrelation(sample_df[calculate_name],1)
    sample_12_lags = autocorrelation(sample_df[calculate_name],12)
    print(calculate_name, sample_mean, sample_std, sample_1_lags, sample_12_lags)
    
    
    
```

```python
# 利差统计
```

```python
import statistics 
calculate_margin_name_list = ['MARGIN_RIFLGFCM06_N.B', 'MARGIN_RIFLGFCY01_N.B', 'MARGIN_RIFLGFCY02_N.B', 'MARGIN_RIFLGFCY03_N.B', 
                       'MARGIN_RIFLGFCY05_N.B', 'MARGIN_RIFLGFCY10_N.B']

N = sample_df.shape[0]

for calculate_margin_name in calculate_margin_name_list:
    sample_mean = sample_df[calculate_margin_name].mean() * 100
    sample_std = sample_df[calculate_margin_name].std() * np.sqrt(N)/np.sqrt(N-1)
    sample_1_lags = autocorrelation(sample_df[calculate_margin_name],1)
    sample_12_lags = autocorrelation(sample_df[calculate_margin_name],12)
    print(calculate_name, sample_mean, sample_std, sample_1_lags, sample_12_lags)
```

```python
# 绘制6， 12， 24， 36， 60， 120月的债券收益率
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(sample_df['Time_Period'], sample_df['RIFLGFCM06_N.B'], label="6M")
ax.plot(sample_df['Time_Period'], sample_df['RIFLGFCY01_N.B'], label="12M")
ax.plot(sample_df['Time_Period'], sample_df['RIFLGFCY02_N.B'], label="24M")
ax.plot(sample_df['Time_Period'], sample_df['RIFLGFCY03_N.B'], label="36M")
ax.plot(sample_df['Time_Period'], sample_df['RIFLGFCY05_N.B'], label="60M")
ax.plot(sample_df['Time_Period'], sample_df['RIFLGFCY10_N.B'], label="120M")

ax.set_title("Bond Yields in Two Dimensions")
ax.set_xlabel("Year")
ax.set_ylabel("Bond Yield")
fig.autofmt_xdate()
plt.tight_layout()
plt.legend(loc=1)
```

```python
# 主成分分析
from sklearn.decomposition import PCA

X = sample_df[['RIFLGFCM06_N.B', 'RIFLGFCY01_N.B', 'RIFLGFCY02_N.B',
                 'RIFLGFCY03_N.B', 'RIFLGFCY05_N.B', 'RIFLGFCY10_N.B']]
pca = PCA(n_components=3)
reduced_X =pca.fit_transform(X)
first_x = []
second_x = []
third_x = []
for i in range(len(reduced_X)):
    first_x.append(reduced_X[i][0])
    second_x.append(reduced_X[i][1])
    third_x.append(reduced_X[i][2])
```

```python


fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(sample_df['Time_Period'], first_x, label="First")
ax.plot(sample_df['Time_Period'], second_x, label="Second")
ax.plot(sample_df['Time_Period'], third_x, label="Third")


ax.set_title("Bond Yield Principal Components")
ax.set_xlabel("Year")
ax.set_ylabel("Principal component")
fig.autofmt_xdate()
plt.tight_layout()
plt.legend(loc=1)
```

```python
# 计算样本标准差，第一阶和第十二阶主成分样本的自相关函数

first_std = np.std(first_x) * np.sqrt(len(first_x))/np.sqrt(len(first_x)-1)
second_std = np.std(second_x) * np.sqrt(len(second_x))/np.sqrt(len(second_x)-1)
third_std = np.std(third_x) * np.sqrt(len(third_x))/np.sqrt(len(third_x)-1)
print(first_std, second_std, third_std)
```

```python
first_1_lags = autocorrelation(first_x, 1)
second_1_lags = autocorrelation(second_x, 1)
third_1_lags = autocorrelation(third_x, 1)
print(first_1_lags, second_1_lags, third_1_lags)
```

```python
first_12_lags = autocorrelation(first_x, 12)
second_12_lags = autocorrelation(second_x, 12)
third_12_lags = autocorrelation(third_x, 12)
print(first_12_lags, second_12_lags, third_12_lags)
```
